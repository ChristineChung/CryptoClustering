```python
# Import required libraries and dependencies
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame and make the index the "coin_id" column.
market_data_df = pd.read_csv("Resources/crypto_market_data.csv", index_col="coin_id")

# Display sample data
market_data_df.head(10)
```

```python
# Generate summary statistics
market_data_df.describe()
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(market_data_df)
```

```python
# Create a DataFrame with the scaled data
scaled_df = pd.DataFrame(market_data_scaled, columns=market_data_df.columns)

# Copy the crypto names from the original data
scaled_df['coin_id'] = market_data_df.index

# Set the coinid column as index
scaled_df = scaled_df.set_index('coin_id')

# Display sample data
scaled_df.head(10)
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k_values = list(range(1, 11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using the scaled DataFrame
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k, random_state=1)
    # 2. Fit the model to the data using the scaled DataFrame
    model.fit(scaled_df)
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k_values,
    "inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

# Display the DataFrame
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_df.plot.line(x="k", y="inertia", title="Elbow Curve", xlabel="Clusters", ylabel="Inertia")
```

#### Answer the following question: 
**Question:** What is the best value for `k`?

**Answer:** The best number of clusters, `k`, can be found by looking at the inertia values, which show how spread out the clusters are. We want to find the "elbow" point where the decrease in inertia slows down. The inertia drops a lot from `k = 1` (287.00) to `k = 2` (198.57) and then to `k = 3` (123.19). After `k = 4` (79.02), the decrease is smaller, meaning the improvement is less significant. So, `k = 4` is the best choice, balancing simplicity and clustering quality.

### Cluster Cryptocurrencies with K-means Using the Original Scaled Data.

```python
# Initialize the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
predicted = kmeans_model.predict(scaled_df)

# View the resulting array of cluster values.
predicted
```

```python
# Create a copy of the DataFrame
clustered_df = scaled_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
clustered_df['Cluster'] = predicted

# Display sample data
clustered_df
```

```python
# Create a scatter plot using Pandas plot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Use "rainbow" for the color to better visualize the data.
scatter_plot = clustered_df.plot.scatter(x="price_change_percentage_24h", y="price_change_percentage_7d", c="Cluster", colormap="rainbow", title="Cryptocurrency Clusters")
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` on the original scaled DataFrame to reduce to three principal components.
pca_data = pca.fit_transform(scaled_df)

# View the first five rows of the DataFrame. 
pca_data[0:5]
```

```python
# Retrieve the explained variance to determine how much information  can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 
Total Percentage/sum of 3 components: 89.5%
Using a special tool called PCA, we turned our big set of data into just three pieces. These pieces are like three big puzzle pieces that show most of the important information from the original data. 

The first piece, PC1, shows about 37 out of every 100 important parts of the puzzle. The second piece, PC2, shows about 35 out of every 100 important parts. The third piece, PC3, shows about 18 out of every 100 important parts. 

When we put these three pieces together, they show almost 90 out of every 100 important parts from the original puzzle. This way, we can understand the big picture without needing all the tiny details.

```python
# Create a new DataFrame with the PCA data.
# Note: The code for this step is provided for you

# Creating a DataFrame with the PCA data
pca_df = pd.DataFrame(pca_data, columns=["PCA1", "PCA2", "PCA3"])

# Copy the crypto names from the original data
pca_df['coin_id'] = market_data_df.index

# Set the coinid column as index
pca_df = pca_df.set_index('coin_id')

# Display sample data
pca_df
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k_values = list(range(1, 11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for k in k_values:

    # 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k, random_state=2)
    # 2. Fit the model to the data using PCA DataFrame
    model.fit(pca_df)
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k_values,
    "inertia": inertia
}
# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

# Display the DataFrame
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_df.plot.line(x="k", y="inertia", title="Elbow Curve", xlabel="Clusters", ylabel="Inertia")
```

#### Answer the following questions: 

**Question:** What is the best number of clusters, `k`, when using PCA data?

**Answer:** To find the best number of clusters, `k`, we look at a special chart called the Elbow curve. This chart shows how spread out the clusters are for different values of `k`. The best `k` is where the curve starts to flatten, meaning adding more clusters doesn't help much. In this case, the best `k` is 4 because that's where the curve flattens out, showing four clusters are just right.

**Question:** Is this different from the best `k` value found using the original data?

**Answer:** No, the best `k` value is 4 in both analysis. Elbow curve seems to be 4 in both analysis. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
kmeans_model.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
predicted = kmeans_model.predict(pca_df)

# Print the resulting array of cluster values.
predicted
```

```python
# Create a copy of the DataFrame with the PCA data
clustered_pca_df = pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
clustered_pca_df['Predicted Cluster'] = predicted

# Display sample data
clustered_pca_df
```

```python
# Create a scatter plot using hvPlot by setting `x="PCA1"` and `y="PCA2"`. 
scatter_plot = clustered_pca_df.plot.scatter(x="PCA1", y="PCA2", c="Predicted Cluster", colormap="winter", title="Cryptocurrency Clusters (PCA)")
```
